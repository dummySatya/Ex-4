```python
# while loops:

c = 10
while c:
    c -= 1
    if(c == 2):
        break
    if(c == 5):
        continue
    print(c) 
else:
    print("While loop has ended")

# The else block only executes when while loop gets terminated by itself
```

```python
# infinite loops
while True:
    print("Bye")
```

```python
# one liners
n = 5
while n > 0 : n -= 1 ; print(n)

# these work only for simple statement, u cant combine other ternary operations inside it
# also its not recommended to use it
```

```python
# for loops

# in python, for loops use iterable objects to iterate

l = ["hello", "this", "is", "me"]

itr = iter(l)

print(itr)
print(next(itr))
print(next(itr))
print(next(itr))
print(next(itr))
print(next(itr))
```

```python
d = {'foo': 1, 'bar': 2, 'baz': 3}

for k in d:
    print(k+ " : ",d[k])

for v in d.values():
    print(v)
```

```python
for i in ['foo', 'bar', 'baz', 'qux']:
    if 'b' in i:
        continue
    print(i)
else:
    print("For loop completed")
```

```python
x = list(range(20,10,-1))

for ind,ele in enumerate(x):
    print(ind,ele)

# enumerate is used to get element and an iterator
```

```python
fruits = ["apple", "banana", "cherry"]
colors = ["red", "yellow", "green"]
for fruit, color in zip(fruits, colors):
    print(fruit, "is", color)
```
